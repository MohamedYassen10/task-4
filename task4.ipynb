```python
#1- facebook and its positive status
facebook=twitter_training_df['Borderlands']=='Facebook'.reset_index()
sns.barplot(y='Borderlands',x='Positive',data=facebook)
plt.show()
```

```python
# Filter the DataFrame to include only rows where 'Borderlands' is 'Facebook'
facebook = twitter_training_df[twitter_training_df['Borderlands'] == 'Facebook']

# Plotting the bar chart using seaborn
sns.lineplot(x='Positive', y='Borderlands', data=facebook)
plt.xlabel("Positive Status")
plt.ylabel("Borderlands")
plt.title("Positive Status for Facebook")
plt.show()
```

```python
# Filter the DataFrame for rows where 'Borderlands' is 'Facebook' and 'Positive' is 'Positive'
facebook_data = twitter_training_df[(twitter_training_df['Borderlands'] == 'Facebook') & (twitter_training_df['Positive'] == 'Positive')]

# Count the number of positive statuses for Facebook
positive_count = facebook_data.shape[0]

# Plotting the bar chart
plt.bar(x=['Facebook'], height=[positive_count])
plt.xlabel("Platform")
plt.ylabel("Count of Positive Status")
plt.title("Count of Positive Status for Facebook")
plt.show()

```
